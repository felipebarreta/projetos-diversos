Giant Puzzle #5 – Blackjack
Zarathustra resolveu investigar uma oportunidade em um cassino quant que oferece variações do jogo de Blackjack. Em uma das variações disponíveis, você adiciona ao baralho 8 cartas I, com o valor de face contabilizado de maneira complexa.

A dinâmica da carta I é como uma exponenciação do complexo i e funciona da seguinte maneira:

Caso você tenha 1 carta I na mão, ela tem valor i, complexo, e não soma ao valor total, que é real, ou seja, 0 pontos;
Com duas cartas I o cenário muda de figura: I*I = i*i = -1, ou seja, você soma -1 pontos;
Com três cartas, voltamos ao cenário complexo: I3 = -i , ou seja, 0 pontos;
Finalmente, com quatro cartas: I4 = 1, ou seja, 1 ponto.
Considerando essa variação, qual o edge mínimo da casa com o jogador seguindo uma estratégia básica idêntica à do dealer, ou seja, escolhendo um valor fixo “soft” para parar de pedir novas cartas?

Extra
Sabemos que repetir a estratégia do dealer não é a melhor opção disponível.

Cite uma estratégia que aumente o valor esperado do jogador!

O ranking daqueles que solucionarem esse puzzle será baseado no valor esperado encontrado pela estratégia sugerida, sendo que aquele que submeteu a solução correta e o maior valor esperado estará no topo! Você pode reenviar sua solução caso encontre uma estratégia mais eficiente.

Obs.: Nesse jogo não é possível dobrar ou splitar e um baralho novo é utilizado a cada rodada. A estratégia do jogador não precisa parar em 17, pode ser qualquer valor. O edge da casa é a vantagem do dealer em relação ao jogador, ou seja, o lucro médio do dealer para cada real gasto pelo apostador – também podendo ser expresso em termos de porcentagem.

Regras do Blackjack
Blackjack é um dos jogos mais populares nos cassinos pelo mundo. O jogo é jogado entre um dealer contra um ou mais jogadores. Cada jogador recebe duas cartas enquanto o dealer apresenta uma carta virada para cima e outra para baixo. Cada carta vale o valor de face, sendo ‘J’=’Q’=’K’= 10 pontos e ‘A’=1 ou 11 pontos. A rodada é definida por cada jogador tomando uma das seguintes decisões:

Hit: pedir mais uma carta;
Stand: parar;
Dobrar: dobrar a aposta por mais uma carta;
Split: caso as duas primeiras cartas possuam o mesmo valor em pontos, é possível separá-las em dois jogos, onde cada uma dará início a uma nova mão. Uma aposta igual a inicial deve ser feita para a segunda mão.
O jogador ganha se a sua mão somar mais pontos que a do dealer, sem ir acima de 21 pontos (burst). No caso de empate, o jogador recebe de volta todo o valor da aposta, exceto quando o dealer tem um “Blackjack”(A e 10/J/Q/K) e o jogador não, nesse caso o jogador perde.

A estratégia do dealer é fixa: hit enquanto somar um total menor que 17 e stand para um total a partir de “soft” 17. Entenda “soft”como uma mão com um A em que ele é contabilizado como 11.

Em caso de vitória do jogador por “Blackjack” (A e 10/J/Q/K), ele recebe 2.5x a aposta, enquanto que em uma vitória simples ele recebe 2x a aposta inicial.

```python
class jogador:
    
    # inicialização da classe
    #--------------------------------------
    def __init__(self, name, soft, baralho="", my_cards="", adv_cards=""):
    
        random.seed(a=None, version=2)
        
        self.name = name
        self.adv_cards = adv_cards
#         self.adv_p = valores[self.adv_card]

        self.soft = soft
        self.baralho = baralho
        
        self.seq_re = ""
        self.seq_img = ""
        self.seq = my_cards
        
        
        for my_card in my_cards:
            if my_card == "I":
                self.seq_img+= my_card
            else:
                self.seq_re += my_card
        
        self.points = self.conta_pontos(self.seq_re, self.seq_img)
    #--------------------------------------
    # setters
    #--------------------------------------
    def setBaralho(self, baralho):
        self.baralho = baralho
    
    # getters
    #--------------------------------------
    def getBaralho(self):
        return self.baralho
    #--------------------------------------
    def getSeq(self):
        return self.seq
    #--------------------------------------
    def getPoints(self):
        return self.points
    #--------------------------------------
    def getSoft(self):
        return self.soft
    #--------------------------------------
    def getAdvCard(self):
        return self.adv_cards
    #--------------------------------------
    def getAdvPoints(self):
        return self.adv_p
    
    # funções de jogo
    #--------------------------------------
    def isBlackjack(self):
        
        seq1 = "ADAJAQAK"
        seq2 = "DAJAQAKA"
        if (len(self.seq) == 2) and (self.seq in seq1 or self.seq in seq2):
            return True
        else:
            return False
            
    #--------------------------------------
    def hit(self):
        i = random.randrange(0,len(self.baralho))
        card = self.baralho[i]
        self.baralho = self.baralho[0:i] + self.baralho[i+1:len(self.baralho)]
        return card
    
    #--------------------------------------
    def conta_pontos(self, seq_re, seq_img):
        
        p_re = 0
        p_img = 0
        
        if seq_img != "":   
            p_img = 1j**len(seq_img)
        
        for card in seq_re:
            
            if card == "A":
                if p_re+p_img.real+11 >= self.soft and p_re+p_img.real+11 <= 21:
                    p_re += 11
                else:
                    p_re += 1
            else:    
                p_re += valores[card]
        
        
        p = p_re + p_img
        
        return p.real
    #--------------------------------------
    
    def play(self):
        
        self.points = self.conta_pontos(self.seq_re, self.seq_img)
        
        while self.points < self.soft:
            card = self.hit()
            
            self.seq += card
            if card == "I":
                self.seq_img += card
            else:
                self.seq_re += card
                
            self.points = self.conta_pontos(self.seq_re, self.seq_img)
            
    #--------------------------------------
    def print(self):
        print("Pontos " + self.name + " = ", self.getPoints())
        print("Sequência "+ self.name + " = ", self.getSeq())
    
    #--------------------------------------
        
```

```python
# função que realiza a rodada:
#------------------------------------
def rodada(dealer, j):
    
    global p_dealer
    global p_jogador
    global aposta

    global lost_hand_count
    global busts_count
    global total_count
    global empate_count
    
    total_count += 1
    
    p_jogador -= aposta
    
    j.play()
    baralho = j.getBaralho()
    dealer.setBaralho(baralho)
    dealer.play()
    
    pd = dealer.getPoints()
    pj = j.getPoints()
    
    if pd > 21:
        pd = -1
    if pj > 21:
        pj = -1
        
#     if pj == -1 and pd == -1:
#         p_jogador += aposta
    if pj == -1:
        p_dealer += aposta
        busts_count += 1
        
    elif pj > pd and j.isBlackjack() == True:
        p_jogador += 2.5* aposta
        p_dealer -= 1.5*aposta
        
#     elif pj > pd and pj <= 21:
    elif pj > pd:
        p_jogador += 2*aposta
        p_dealer -= aposta
        
    elif pj == pd:
        if dealer.isBlackjack() == True and j.isBlackjack() == False:
            p_dealer += aposta
            lost_hand_count +=1
        else:
            p_jogador += aposta
            empate_count += 1
    
#     elif pd >= pj:
#         p_dealer += aposta
    
    else:
        p_dealer += aposta
        lost_hand_count +=1

#------------------------------------  
def hit(baralho):
    random.seed(a=None, version=2)
    i = random.randrange(0,len(baralho))
    card = baralho[i]
    baralho = baralho[0:i] + baralho[i+1:len(baralho)]
#         print("i = ", i)
#         print("baralho = ", baralho)
    return card, baralho
#------------------------------------  
def select(baralho, card):
    baralho = baralho[0:baralho == card] + baralho[(baralho == card)+1: len(baralho)]
    return card, baralho

#------------------------------------    
# função que realiza o jogo completo
def game(p, n_max, soft_dealer, soft_jogador, carta_dealer):

    global p_dealer
    global p_jogador
    global aposta
    
    global lost_hand_count
    global busts_count
    global total_count
    global empate_count
    
    lost_hand_count = 0
    busts_count = 0
    total_count = 0
    empate_count = 0
    
    p_dealer_ini = p
    p_dealer = p

    p_jogador_ini = p
    p_jogador = p

    for i in range(n_max):


        baralho = "AAAA22223333444455556666777788889999DDDDJJJJQQQQKKKKIIIIIIII"
    #     baralho = "AAAA22223333444455556666777788889999DDDDJJJJQQQQKKKK"

        dealer_card1, baralho = select(baralho, carta_dealer)
        dealer_card2, baralho = hit(baralho)

        dealer_cards = dealer_card1 + dealer_card2

        j1_card1, baralho = hit(baralho)
        j1_card2, baralho = hit(baralho)

        j1_cards = j1_card1 + j1_card2    

        dealer = jogador("dealer", soft_dealer, baralho, dealer_cards, j1_cards)
        j1 = jogador("jogador", soft_jogador, baralho, j1_cards, dealer_cards)
        rodada(dealer, j1)


    # dealer.print()
    # j1.print()
    edge = 100 * (p_dealer - p_dealer_ini) / (n_max*aposta)
    
#     print("Resultado com soft joagador = ", soft_jogador, "e carta do dealer = ", carta_dealer)
#     print("Quantidade de iterações = ", n_max)
#     print("Dinheiro do dealer = ", p_dealer)
#     print("Dinheiro do jogador = ", p_jogador)
#     print("Edge do dealer = ", edge, "%")
    # print("baralho = ", j1.getBaralho())
    # print("len baralho = ", len(j1.getBaralho()))
    
    return edge

#------------------------------------

p = 1e6
n_max = 10000
aposta = 1
min_soft_jogador = 10
soft_dealer = 17
list_soft_jogador = list(range(min_soft_jogador,22))

cartas = "A23456789DJQKI"

results = np.array(np.zeros([len(cartas), len(list_soft_jogador)]))
lost_hand = np.array(np.zeros([len(cartas), len(list_soft_jogador)]))
busts = np.array(np.zeros([len(cartas), len(list_soft_jogador)]))
empate = np.array(np.zeros([len(cartas), len(list_soft_jogador)]))
total = np.array(np.zeros([len(cartas), len(list_soft_jogador)]))

# posicao = {"A": 0, "2": 1, "3": 2, "4": 3, "5": 4, "6": 5, "7": 6, "8": 7, "9": 8, "D": 9, "J": 9, "Q": 9, "K": 9, "I": 10}

for carta_dealer in cartas:
    for soft_jogador in list_soft_jogador:
        edge = game(p, n_max, soft_dealer, soft_jogador, carta_dealer)
        results[cartas.index(carta_dealer)][soft_jogador - min_soft_jogador]= edge
        lost_hand[cartas.index(carta_dealer)][soft_jogador - min_soft_jogador]= lost_hand_count
        busts[cartas.index(carta_dealer)][soft_jogador - min_soft_jogador]= busts_count
        empate[cartas.index(carta_dealer)][soft_jogador - min_soft_jogador]= empate_count
        total[cartas.index(carta_dealer)][soft_jogador - min_soft_jogador]= total_count
        

print("Quantidade de iterações = ", n_max)
np.set_printoptions(precision=3)

df_results = pd.DataFrame(results,index=list(cartas), columns=list_soft_jogador)
print("Edge do dealer")
display(df_results)

df_total = pd.DataFrame(total,index=list(cartas), columns=list_soft_jogador)
print("Total")
display(df_total)

df_busts = pd.DataFrame(100*busts/total,index=list(cartas), columns=list_soft_jogador)
print("Busts")
display(df_busts)

df_lost_hand = pd.DataFrame(100*lost_hand/total,index=list(cartas), columns=list_soft_jogador)
print("Perdas")
display(df_lost_hand)

df_empate = pd.DataFrame(100*empate/total,index=list(cartas), columns=list_soft_jogador)
print("Empate")
display(df_empate)

df_lost_hand_plus_busts = pd.DataFrame(100*(lost_hand/total + busts / total),index=list(cartas), columns=list_soft_jogador)
print("Perdas + Busts")
display(df_lost_hand_plus_busts)
```
